DISTRICT SUMMARY

```python
#Imports modules pandas and pathlib
import pandas as pd
from pathlib import Path

#Creates path to file
path_to_school_data = Path("Resources/schools_complete.csv")
path_to_student_data = Path("Resources/students_complete.csv")

#Reads the csv
school_data = pd.read_csv(path_to_school_data)
student_data = pd.read_csv(path_to_student_data)

#Combines the data into a single dataset using pd.merge(csv1, csv2, ?, on school name)  
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(10)
```

```python
# Calculate the total number of unique schools
#this would use value counts to check the number of instances
school_count = len(school_data_complete["school_name"].value_counts())
school_count

print("There are", school_count, "schools")
```

```python
# Calculate the total number of students
#looking for total, use sum function
#did value_counts() to find how many students per school, then sum() to total it together
student_count = school_data_complete["school_name"].value_counts().sum()
student_count

print("There are", student_count, "students total")
```

```python
# Calculate the total budget
#SO looked for each unique value in budget
#e.g. huang hs has one budget '1,910,635'
#find unique instances of it (how many budgets there are) and return them
#use sum function to the array of budgets together ([5, 5, 5, 10]).sum() would return 25
total_budget = (school_data_complete["budget"].unique()).sum()
total_budget
print("Total budget is", total_budget)
```

```python
# Calculate the average (mean) math score
average_math_score = (school_data_complete["math_score"].sum())/student_count
average_math_score

print("The average math score is", average_math_score)
```

```python
# Calculate the average (mean) reading score
average_reading_score = (school_data_complete["reading_score"].sum())/student_count
average_reading_score

print("The average reading score is", average_reading_score)
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage

print("Percent of students who passed math is", passing_math_percentage)
#this code was given to me
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage

print("Percent of students who passed reading is", passing_reading_percentage)
#copied and pasted
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate

print("The percent of students who passed math and reading is", overall_passing_rate)
#this code was given to me
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    "Total Schools" : [school_count],
    "Total Students" : [student_count],
    "Total Budget" : [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score" : [average_reading_score],
    "% Passing Math" : [passing_math_percentage],
    "% Passing Reading" : [passing_reading_percentage],
    "% Overall Passing" : [overall_passing_rate]
})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

SCHOOL SUMMARY

```python
#Sets up a dataframe with school name, type, and budget
school_summary_df = school_data_complete[["school_name", "type", "budget"]] 

#drops duplicate index values in column (subset) "school_name"
school_summary_df = school_summary_df.drop_duplicates(subset = "school_name")

#Sorts the names alphabetically
school_summary_df = school_summary_df.sort_values("school_name")

#Changes the index to "school_name"
school_summary_df = school_summary_df.set_index("school_name")

#Displays the dataframe 
school_summary_df
```

```python
#counts how many times each school's name appears in the "school_name" column -> counts students per school
stu_per_school = school_data_complete["school_name"].value_counts()

#turns the series into a dictionary e.g. "Bailey HS" : "365"
stu_per_school.to_dict()


#Inserts the column as "students" and with the dictionary, uses the index keys to put the values in the right spot
school_summary_df["Students"] = stu_per_school 

#Displays the dataframe 
school_summary_df
```

```python
#gathers total budget as a series
tot_budget = school_summary_df["budget"]

#gathers students per school as a series
num_of_stu = school_summary_df["Students"]

#finds budget per student as total_budget / num_of_stu
bug_per_stu = tot_budget / num_of_stu

#inserts bug_per_stu under column "Budget per Student"
school_summary_df["Budget per Student"] = bug_per_stu

#Displays the dataframe 
school_summary_df
```

```python
#need average test scores... a new dataframe with scores and school... and then what? 
#could do a sum of the scores in the original df and then divide by number of students
#add it in
scores_df = school_data_complete[["school_name", "reading_score", "math_score"]]

#total scores equals scores dataframe and group by school name
tot_scores = scores_df.groupby(["school_name"])

#sums up all scores by school -> then can divide by number of students in each school
tot_scores_two = tot_scores.sum()

tot_scores_two

reading_series = pd.Series(tot_scores_two["reading_score"])
av_read_score = reading_series / num_of_stu
school_summary_df["Average Reading Score"] = av_read_score

math_series = pd.Series(tot_scores_two["math_score"])
av_math_score = math_series / num_of_stu
av_math_score

school_summary_df["Av Math Score"] = av_math_score

school_summary_df
```

```python
#uses query to filter out scores >= 70 but keeps all scores individually -> huang hs 70, huanghs 91, huanghs 82, etc...
pass_math_scores = scores_df.query("math_score >= 70")
#groups the scores by school but groupby returns a string
pass_math_gb_str = pass_math_scores.groupby("school_name")
#groups and saves to dataframe the scores by school so its total scores per school -> huanghs 452 (452 scores >= 70)
pass_math_gb = pass_math_gb_str.count()
#turns math scores into a series
pass_math_series = pd.Series(pass_math_gb["math_score"])
#calculates percent who passed
perc_pass_math = (pass_math_series / num_of_stu)*100
#displays result
perc_pass_math

school_summary_df["% Passing Math"] = perc_pass_math

#uses query to filter out scores >= 70 but keeps all scores individually -> huang hs 70, huanghs 91, huanghs 82, etc...
pass_read_scores = scores_df.query("reading_score >= 70")
#groups the scores by school but groupby returns a string
pass_read_gb_str = pass_read_scores.groupby("school_name")
#groups and saves to dataframe the scores by school so its total scores per school -> huanghs 452 (452 scores >= 70)
pass_read_gb = pass_read_gb_str.count()
#turns reading scores into a series
pass_read_series = pd.Series(pass_read_gb["reading_score"])
#calculates percent who passed
per_pass_read = (pass_read_series / num_of_stu)*100
#displays result
per_pass_read

school_summary_df["% Passing Reading"] = per_pass_read

#uses query to filter out ALL scores >= 70
pass_both_scores = scores_df.query("math_score >= 70 & reading_score >= 70")
#groups the scores by school but groupby returns a string
pass_both_gb_str = pass_both_scores.groupby("school_name")
#groups and saves to dataframe the scores by school so its total scores per school 
pass_both_gb = pass_both_gb_str.count()
#turns scores into a series, #both columns have same numbers
pass_both_series = pd.Series(pass_both_gb["reading_score"]) 
#calculates percent who passed
per_pass_both = (pass_both_series / num_of_stu)*100
#displays result
per_pass_both

#school_summary_df["% Passing Math"] = perc_pass_math
school_summary_df["% Overall Passing"] = per_pass_both

school_summary_df
```

```python
#reorders columns
school_summary_df = school_summary_df[["type", "Students", "budget","Budget per Student", "Av Math Score", 
                                       "Average Reading Score","% Passing Math", "% Passing Reading", "% Overall Passing"]]
school_summary_df
```

```python
#rename columns
school_summary_df = school_summary_df.rename(columns= {
    "type" : "School Type", 
    "Students": "Total Students",
    "budget" : "Total School Budget", 
    "Budget per Student" : "Per Student Budget", 
    "Av Math Score" : "Average Math Score"
})
school_summary_df
```

SCORES BY SCHOOL SPENDING

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labelsz = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = school_summary_df.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(bug_per_stu, spending_bins, 
                                                                                labels= labelsz, include_lowest = True)
school_spending_df
```

```python
spending_summary = school_spending_df[["Spending Ranges (Per Student)", "Average Math Score", "Average Reading Score", 
                                       "% Passing Math","% Passing Reading", "% Overall Passing" ]]
spending_summary = spending_summary.reset_index(drop = True)
spending_sum_df = spending_summary.groupby("Spending Ranges (Per Student)")
spending_sum_df.mean()
```
